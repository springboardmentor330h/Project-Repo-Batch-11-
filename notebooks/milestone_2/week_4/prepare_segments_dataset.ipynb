**IMPORTS**

```python
import streamlit as st    # Import Streamlit for building the web app
import json
import os
import pandas as pd
from pathlib import Path    # Import Path to work with file paths easily
```

**PATHS**

```python
BASE_PATH   = "/content/drive/MyDrive/podcast-project"
SEGMENT_DIR = f"{BASE_PATH}/data/segmented_outputs"
AUDIO_DIR   = f"{BASE_PATH}/data/audio_raw"
```

**CREATE COMBINED SEGMENTS CSV**

```python
# Creates combined CSV

# Create an empty list to store all segment data
all_data = []

# Loop through all files in the segment directory
for file in os.listdir(SEGMENT_DIR):
    if file.endswith("_segment.json"):

        # Open and read the JSON file
        with open(os.path.join(SEGMENT_DIR, file), 'r') as f:
            ep = json.load(f)
            ep_id = ep["episode_id"]

            # Loop through each segment in the episode
            for seg in ep["segments"]:

                # Add the segment details to the list
                all_data.append({
                    "episode_id": ep_id,
                    "segment_id": seg["segment_id"],
                    "summary": seg["summary"],
                    "keywords": ", ".join(seg.get("keywords", [])),
                    "start_time_sec": seg.get("start_time_sec", 0.0),
                    "text_preview": seg["text_preview"],
                    "num_sentences": seg["num_sentences"]
                })

df_all = pd.DataFrame(all_data)   # Convert the list of data into a Pandas DataFrame
df_all.to_csv("/content/drive/MyDrive/podcast-project/data/segments_processed/all_segments.csv", index=False)
print("Saved combined file – ready to use this in app")
```
