```python
import os
import librosa
import soundfile as sf
import noisereduce as nr

OUTPUT_DIR = "/content/drive/MyDrive/processed_audio"
os.makedirs(OUTPUT_DIR, exist_ok=True)

for file in os.listdir(INPUT_DIR):
    if file.lower().endswith(".mp3"):
        audio, sr = librosa.load(os.path.join(INPUT_DIR, file), sr=16000, mono=True)
        reduced_audio = nr.reduce_noise(y=audio, sr=sr)
        sf.write(
            os.path.join(OUTPUT_DIR, file.replace(".mp3", ".wav")),
            reduced_audio,
            sr
        )

print("✅ Audio preprocessing completed")
```

```python
from pydub import AudioSegment
import os

INPUT_DIR = "/content/drive/MyDrive/processed_audio"
OUTPUT_DIR = "/content/drive/MyDrive/audio_chunks"

os.makedirs(OUTPUT_DIR, exist_ok=True)

CHUNK_LENGTH_MS = 25 * 1000  # 25 seconds

for file in os.listdir(INPUT_DIR):
    if file.endswith(".wav"):
        audio = AudioSegment.from_wav(os.path.join(INPUT_DIR, file))
        base_name = file.replace(".wav", "")

        for i in range(0, len(audio), CHUNK_LENGTH_MS):
            chunk = audio[i:i + CHUNK_LENGTH_MS]
            chunk_name = f"{base_name}_chunk{i//CHUNK_LENGTH_MS}.wav"
            chunk.export(os.path.join(OUTPUT_DIR, chunk_name), format="wav")

print("✅ Audio chunking completed")
```

```python
import os
import json

TRANSCRIPTS_DIR = "/content/drive/MyDrive/transcripts"

full_text = ""

for file in sorted(os.listdir(TRANSCRIPTS_DIR)):
    if file.endswith(".json"):
        with open(os.path.join(TRANSCRIPTS_DIR, file), "r") as f:
            data = json.load(f)
            full_text += data["text"] + " "

print(full_text[:500])  # preview
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(full_text)
print(sentences[:5])
print("Total sentences:", len(sentences))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(sentences)
```

```python
print(X.shape)
```

```python
threshold = 0.2

segments_algo1 = []
current_segment = [sentences[0]]

for i in range(len(sentences) - 1):
    sim = cosine_similarity(X[i], X[i+1])[0][0]

    if sim < threshold:
        segments_algo1.append(current_segment)
        current_segment = [sentences[i+1]]
    else:
        current_segment.append(sentences[i+1])

segments_algo1.append(current_segment)

print("Total topics (Algorithm 1):", len(segments_algo1))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(sentences)

segments_algo2 = []
current_segment = [sentences[0]]

for i in range(len(sentences) - 1):
    sim = cosine_similarity(
        [embeddings[i]], [embeddings[i+1]]
    )[0][0]

    if sim < 0.5:
        segments_algo2.append(current_segment)
        current_segment = [sentences[i+1]]
    else:
        current_segment.append(sentences[i+1])

segments_algo2.append(current_segment)

print("Total topics (Algorithm 2):", len(segments_algo2))
```

```python
final_segments = segments_algo2
```

```python
from collections import Counter
import re

def extract_keywords(segment, top_n=8):
    text = " ".join(segment).lower()
    words = re.findall(r"\b[a-z]+\b", text)
    return Counter(words).most_common(top_n)

for i, seg in enumerate(final_segments[:3]):
    print(f"Topic {i+1} keywords:", extract_keywords(seg))
```

```python
for i, seg in enumerate(final_segments[:3]):
    print(f"\nTopic {i+1} summary:")
    print(" ".join(seg[:2]))
```
